```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load("combined_processed.npz", allow_pickle=True)
normal = data["normal"]
relay = data["relay"]

print(f"Normal shape: {normal.shape}, Relay shape: {relay.shape}")

# Einzelne Zeitreihe plotten
idx = 0  # Index der Zeitreihe
plt.figure(figsize=(10,4))
plt.plot(normal[idx], label=f"Normal[{idx}]")
plt.xlabel("Zeit (Samples)")
plt.ylabel("Signalwert")
plt.title(f"Normal Zeitreihe {idx}")
plt.legend()
plt.show()

# Mehrere Zeitreihen nebeneinander
plt.figure(figsize=(12,6))
for i in range(5000,5500):  # die ersten 5 Normal-Samples
    plt.plot(normal[i], alpha=0.7, label=f"Normal[{i}]")
plt.xlabel("Zeit (Samples)")
plt.ylabel("Signalwert")
plt.title("Mehrere Normal-Zeitreihen")
plt.legend()
plt.show()

# Vergleich Normal vs Relay
plt.figure(figsize=(12,6))
plt.plot(normal[0], label="Normal[0]", alpha=0.8)
plt.plot(relay[0], label="Relay[0]", alpha=0.8)
plt.xlabel("Zeit (Samples)")
plt.ylabel("Signalwert")
plt.title("Vergleich Normal vs Relay")
plt.legend()
plt.show()
```
